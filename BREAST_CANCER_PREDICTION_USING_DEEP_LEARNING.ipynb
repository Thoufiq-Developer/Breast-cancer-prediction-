```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
a=load_breast_cancer()
data=pd.DataFrame(a.data,columns=a.feature_names)
data.head()
```

```python
print(a)
```

```python
data['lables']=a.target
```

```python
X=data.drop(columns='lables',axis=1)
```

```python
Y=data.lables
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=2,test_size=0.2)
```

```python
print(X.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

STANDARDIZE THE DATA

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
a=scaler.fit_transform(x_train)
b=scaler.transform(x_test)
```

NEURAL NETWORK CREATION


```python
import tensorflow as tf
tf.random.set_seed=3
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(
           optimizer='adam',
           loss='sparse_categorical_crossentropy',
           metrics=['accuracy']
)
```

```python
history=model.fit(x_train,y_train,epochs=10,validation_split=0.1)
```

VISUALIZING THE ACCURACY AND LOSS


```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('accuracy prediction')
plt.ylabel=('accuracy')
plt.xlabel=('epochs')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('accuracy prediction')
plt.ylabel=('loss')
plt.xlabel=('epochs')

plt.legend(['training data','validation data'],loc='upper right')
```

Checking with the test data


```python
loss, accuracy=model.evaluate(b,y_test)
print(accuracy)
```

```python
input_data=(12.9,15.92,83.74,512.2,0.08677,0.09509,0.04894,0.03088,0.1778,0.06235,0.2143,0.7712,1.689,16.64,0.005324,0.01563,0.0151,0.007584,0.02104,0.001887,14.48,21.82,97.17,643.8,0.1312,0.2548,0.209,0.1012,0.3549,0.08118)
input_datas=np.asarray(input_data)
reshaping=input_datas.reshape(1,-1)
standardized_data=scaler.transform(reshaping)
prediction=model.predict(standardized_data)
final_result=np.argmax(prediction)
print(final_result)
if final_result==0:
  print('The breast cancer is Malignant')
else:
  print('The breast cancer is Benign')
```
